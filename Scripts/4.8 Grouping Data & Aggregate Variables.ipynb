# loyalty_flag

```python
# create 'max_order' variable
orders_products_flagged['max_order'] = orders_products_flagged.groupby(['customer_id'])['order_number'].transform(np.max)
```

```python
# loyalty flag
orders_products_flagged.loc[orders_products_flagged['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
orders_products_flagged.loc[(orders_products_flagged['max_order'] <= 40) & (orders_products_flagged['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
orders_products_flagged.loc[orders_products_flagged['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# frequency of 'loyalty_flag'
orders_products_flagged['loyalty_flag'].value_counts(dropna = False)
```

# loyalty_flag

```python
# get min, max, mean of 'prices' gouped by 'loyalty_flag'
orders_products_flagged.groupby('loyalty_flag').agg({'prices':['mean','min','max']})
```

# spending_flag

```python
# create 'mean_order' variable
orders_products_flagged['mean_order'] = orders_products_flagged.groupby(['customer_id'])['order_number'].transform(np.mean)
```

```python
# create flag 'spending_flag'
orders_products_flagged.loc[orders_products_flagged['mean_order']<10, 'spending_flag'] = 'Low spender'
orders_products_flagged.loc[orders_products_flagged['mean_order']>=10, 'spending_flag'] = 'High spender'
```

```python
# check
orders_products_flagged.head()
```

```python
# frequency of 'spending_flag'
orders_products_flagged['spending_flag'].value_counts(dropna=False)
```

# order_frequency_flag

```python
# create median 'days_since_prior_order_median'
orders_products_flagged['days_since_prior_order_median'] = orders_products_flagged.groupby(['customer_id'])['days_since_prior_order'].transform(np.median)
```

```python
# create flag 'order_frequency_flag'
orders_products_flagged.loc[orders_products_flagged['days_since_prior_order_median']>20, 'order_frequency_flag'] = 'Non-frequent customer'
orders_products_flagged.loc[orders_products_flagged['days_since_prior_order_median']<=20, 'order_frequency_flag'] = 'Regular customer'
orders_products_flagged.loc[orders_products_flagged['days_since_prior_order_median']<=10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
# frequency of 'order_frequency_flag'
orders_products_flagged['order_frequency_flag'].value_counts(dropna=False)
```

```python
# check those 5 NaNs
orders_products_flagged[orders_products_flagged['order_frequency_flag'].isnull()]
```

```python
# check
orders_products_flagged.head(50)
```

```python
# deleting those 5 NaNs as they're useless and that's okay because it won't affect the whole dataset
orders_products_flagged = orders_products_flagged[orders_products_flagged['order_frequency_flag'].notnull()]
```
