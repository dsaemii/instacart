# Region

```python
# create lists with states for each region
northeast = ['Maine', 'New Hampshire', 'Vermont', 'Massachusetts', 'Rhode Island', 'Connecticut', 'New York', 'Pennsylvania', 'New Jersey']
```

```python
midwest = ['Wisconsin', 'Michigan', 'Illinois', 'Indiana', 'Ohio', 'North Dakota', 'South Dakota', 'Nebraska', 'Kansas', 'Minnesota', 'Iowa', 'Missouri']
```

```python
south = ['Delaware', 'Maryland', 'District of Columbia', 'Virginia', 'West Virginia', 'North Carolina', 'South Carolina', 'Georgia', 'Florida', 'Kentucky', 'Tennessee', 'Mississippi', 'Alabama', 'Oklahoma', 'Texas', 'Arkansas', 'Louisiana']
```

```python
west = ['Idaho', 'Montana', 'Wyoming', 'Nevada', 'Utah', 'Colorado', 'Arizona', 'New Mexico', 'Alaska', 'Washington', 'Oregon', 'California', 'Hawaii']
```

```python
# create region
orders_product_all.loc[orders_product_all['state'].isin(northeast), 'region'] = 'northeast'
```

```python
orders_product_all.loc[orders_product_all['state'].isin(midwest), 'region'] = 'midwest'
```

```python
orders_product_all.loc[orders_product_all['state'].isin(south), 'region'] = 'south'
```

```python
orders_product_all.loc[orders_product_all['state'].isin(west), 'region'] = 'west'
```

```python
# check
orders_product_all['region'].value_counts(dropna = False)
```

```python
# spending habits in US regions
# northeast
orders_product_all.loc[orders_product_all['region'] == 'northeast']['spending_flag'].value_counts().plot.bar()
```

```python
# northeast
orders_product_all.loc[orders_product_all['region'] == 'midwest']['spending_flag'].value_counts().plot.bar()
```

```python
# northeast
orders_product_all.loc[orders_product_all['region'] == 'south']['spending_flag'].value_counts().plot.bar()
```

```python
# northeast
orders_product_all.loc[orders_product_all['region'] == 'west']['spending_flag'].value_counts().plot.bar()
```

###### It looks like there aren't big differences among the regions of the US in spending habits. In general, there are more High spender than Low spender.

```python
# low-activity customers
# create 'orders_count'
orders_product_all['orders_count'] = orders_product_all.groupby('customer_id')['customer_id'].transform('count')
```

```python
# create flag
orders_product_all.loc[orders_product_all['orders_count'] < 5, 'low_activity'] = 'low_activity customer'
```

```python
# check frequency of flag
orders_product_all['low_activity'].value_counts(dropna=False)
```

```python
# delete the low activity customers
orders_product_all = orders_product_all[orders_product_all['low_activity'] != 'low_activity customer']
```
